<h4>Another Example of using Principal Component Analysis</h4>

```
 import numpy as np
```

```
 X = np.array([[10,20,10],              [2,5,2],              [8,17,7],              [9,20,10],              [12,22,11]])
```

```
 print X
```

```
 X = np.mat(X)meanVals = mean(X, axis=0)A = X - meanVals              # A is the zero-mean (centered) version of XC = np.cov(A, rowvar=0)       # C is the covarianvce matrix of X print C
```

```
 # Note that C = (1/(N-1)) A.T*Aprint dot(A.T,A)/(shape(X)[0]-1)
```

<h4>Now we can obtain eigenvalues and eigenvectors of the covariance matrix:</h4>

```
 np.set_printoptions(precision=2,suppress=True)e, ev = eig(C)print "Eigenvalues:"print eprintprint "Eigenvectors:"print ev
```

<h4>We can transform the full data into the new feature space based on the eigenvectors:</h4>

```
 newFeatures = ev.TXTrans = dot(newFeatures, A.T)print XTrans.T
```

<h4>However,
 typically, we want a lower-dimensional space. We can sort the 
eigenvectors in the decreasing order of their eigenvalues and take the 
top k. In the example below, we'll take only the top first principal 
component (since it has the largest eigenvalue, no sorting necessary):</h4>

```
 reducedFeatures = ev[:,0].TredcuedXTrans = dot(reducedFeatures, A.T)print redcuedXTrans.T
```

<h4>We can also use Scikit-learn decomposition module to do the same thing:</h4>

```
 from sklearn import decomposition
```

```
 pca = decomposition.PCA()XTrans2 = pca.fit_transform(X)
```

```
 print XTrans2
```
