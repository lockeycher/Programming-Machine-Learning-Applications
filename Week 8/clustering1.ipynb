
<h4 id="We-will-use-Boston-Housing-data-to-illustrate-the-use-of-clustering-technqiues.">We will use Boston Housing data to illustrate the use of clustering technqiues.<a class="anchor-link" href="https://nbviewer.jupyter.org/url/bmobasher.com/Class/CSC478/Clustering.ipynb#We-will-use-Boston-Housing-data-to-illustrate-the-use-of-clustering-technqiues.">¶</a></h4>


```

import numpy as np
import pylab as pl
import pandas as pd
from sklearn.cluster import KMeans 
from sklearn.datasets import load_boston
boston = load_boston()

```

```

np.set_printoptions(suppress=True, precision=2, linewidth=120)

```

```

print boston.data[:2]

```

```

print boston.target[:2]

```


<h4 id="There-is-no-target-attribute-in-unsupervised-approaches-to-data-mining.-So,-we-can-add-back-the-target-attribute-(in-this-case-it-may-be-a-useful-variable-for-clustering).">There is no target attribute in unsupervised approaches to data mining. So, we can add back the target attribute (in this case it may be a useful variable for clustering).<a class="anchor-link" href="https://nbviewer.jupyter.org/url/bmobasher.com/Class/CSC478/Clustering.ipynb#There-is-no-target-attribute-in-unsupervised-approaches-to-data-mining.-So,-we-can-add-back-the-target-attribute-(in-this-case-it-may-be-a-useful-variable-for-clustering).">¶</a></h4>


```

x = np.array([np.concatenate((boston.data[i],[boston.target[i]])) for i in range(len(boston.data))])

```

```

print x[:2]

```


<h4 id="Now-we-use-KMeans-algorithm-of-scikit-learn-to-perform-the-clustering.">Now we use KMeans algorithm of scikit-learn to perform the clustering.<a class="anchor-link" href="https://nbviewer.jupyter.org/url/bmobasher.com/Class/CSC478/Clustering.ipynb#Now-we-use-KMeans-algorithm-of-scikit-learn-to-perform-the-clustering.">¶</a></h4>


```

kmeans = KMeans(n_clusters=5, max_iter=500, verbose=1) # initialization

```

```

kmeans.fit(x)

```

```

clusters = kmeans.predict(x)

```

```

print clusters

```


<h4 id="The-centroids-provide-an-aggregate-representation-and-a-characterization-of-each-cluster.">The centroids provide an aggregate representation and a characterization of each cluster.<a class="anchor-link" href="https://nbviewer.jupyter.org/url/bmobasher.com/Class/CSC478/Clustering.ipynb#The-centroids-provide-an-aggregate-representation-and-a-characterization-of-each-cluster.">¶</a></h4>


```

print boston.feature_names
print kmeans.cluster_centers_

```

```

pl.plot(data[target==0,0],data[target==0,2],'bo')
pl.plot(data[target==1,0],data[target==1,2],'ro')
pl.plot(data[target==2,0],data[target==2,2],'go')
pl.legend(('Iris-Setosa', 'Iris-Versicolour', 'Iris-Virginica'), loc=4)
pl.show()

```


<h4 id="In-the-graph-we-have-150-points-and-their-color-represents-the-class;-the-blue-points-represent-the-samples-that-belong-to-the-specie-setosa,-the-red-ones-represent-versicolor-and-the-green-ones-represent-virginica.-Next-let's-see-if-through-clustering-we-can-obtain-the-correct-classes.">In the graph we have 150 points and their color represents the class; the blue points represent the samples that belong to the specie setosa, the red ones represent versicolor and the green ones represent virginica. ```

```
iris_kmeans = KMeans(n_clusters=3, max_iter=500, verbose=1, n_init=5) # initialization
iris_kmeans.fit(data)

```

```

c = iris_kmeans.predict(data)

```

<h4 id="Homogeneity:-each-cluster-contains-only-members-of-a-single-class.-Completeness:-all-members-of-a-given-class-are-assigned-to-the-same-cluster.">Homogeneity: each cluster contains only members of a single class. Completeness: all members of a given class are assigned to the same cluster.<a class="anchor-link" href="https://nbviewer.jupyter.org/url/bmobasher.com/Class/CSC478/Clustering.ipynb#Homogeneity:-each-cluster-contains-only-members-of-a-single-class.-Completeness:-all-members-of-a-given-class-are-assigned-to-the-same-cluster.">¶</a></h4>


```

from sklearn.metrics import completeness_score, homogeneity_score

```

```

print completeness_score(target,c)

```

```

print homogeneity_score(target,c)

```
